```
data_num.info()
# as we can see our dataframe is missing some values and also containts some questionable variables, let's take care of it 
```

```
# dropping the dollar sign and transorming it to floats 
data_num['cleaning_fee'] = data_num['cleaning_fee'].str.replace('$','').str.replace(',','').astype(float)
data_num['price'] = data_num['price'].str.replace('$','').str.replace(',','').astype(float)
data_num['security_deposit'] = data_num['security_deposit'].str.replace('$','').str.replace(',','').astype(float)
data_num['extra_people'] = data_num['extra_people'].str.replace('$','').str.replace(',','').astype(float)
```

```
# too many observations are missing data in 'weekly_price', 'monthly_price' and 'square_feet' columns, therefore we will drop these columns
data_num = data_num.drop(columns = ['weekly_price','monthly_price','square_feet'])
data_num.info()
```

```
# let's assume that the missing data for the 'securyity deposit' and 'cleaning_fee' is missing because it is equal to 0 
data_num['security_deposit'] = data_num['security_deposit'].fillna(0)
data_num['cleaning_fee'] = data_num['cleaning_fee'].fillna(0)
```

```
# since we still have missing data let's impute it (make a calculated guess about the missing values). For this I use K-Nearest_neighbour Classifier from the fancyimpute package
knn_imputer = KNN()
data_num.iloc[:,:]= knn_imputer.fit_transform(data_num)
```

```
# since our numeric values have various standard deviations and are different in nature we need to bring them to a more or less homogenous form, for this we use the StandardScaler that scales every columns relative to its mean and standard deviation 
scaler = StandardScaler()
scaler.fit(data_num)
data_num_scaled = scaler.transform(data_num)
```

```
ks = range(1, 15) # number of clusters to test 
inertias = [] 

for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters = k)
    
    # Fit model to samples
    model.fit(data_num)
    
    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
    
# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.savefig('intertias linegraph with extra variables.png')

plt.show()

# the output provides a linegraph of inertias given different numbers of clusters
# this is useful to identify the best number of clusters to choose (using the elbow method)
```

```
#splitting the data into dependent and independent variables 
X = data_num_no_outliers.drop(columns= 'price')
y = data_num_no_outliers['price']
```

```
# Using KMeans clustering to split the data into 4 clusters 
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
y_pred = kmeans.predict(X)
y_pred
```

```
# checking how the clusters sorrespond to the relationship between the price and maximum people a listing can accomodate 
fig, ax = plt.subplots(figsize=(16,8))
sns.scatterplot(x=X['accommodates'],y = y, hue = y_pred)
plt.savefig('accomodates ~ price ~ colored.png')
```

```
# let's reduce the dimentionality of our data so we can plot it 
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
print('Original Shape {}'.format(X.shape))
print('Transformed Shape {}'.format(X_pca.shape))
```

```
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
fig, ax = plt.subplots(figsize=(16,8))
ax = fig.add_subplot(111, projection='3d')
# Plot the compressed data points
ax.scatter(X_pca[:,1], y, X_pca[:,0], c=y_pred, s=60)
ax.view_init(30, 185)
ax.set_xlabel('PCA Variable #1')
ax.set_ylabel('Price')
ax.set_zlabel('PCA Variable #2')
plt.savefig('pca 3d colored.png')
plt.show()
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
```

```
# implementing the most basic linear regression model 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Create the regressor: reg_all
reg_all = LinearRegression()

# Fit the regressor to the training data
reg_all.fit(X_train,y_train)

# Predict on the test data: y_pred
y_pred = reg_all.predict(X_test)

# Compute and print R^2 and RMSE
print("R^2: {}".format(reg_all.score(X_test, y_test)))
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
print("Root Mean Squared Error: {}".format(rmse))
```
