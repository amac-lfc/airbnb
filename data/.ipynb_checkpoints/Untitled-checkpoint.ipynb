```python
import pandas as pd
```

```python
data_all = pd.read_csv('listings.csv',index_col=['id'])
data = pd.read_csv('listings_cleaned_with_counts.csv', index_col = ['id'])

cols = ['latitude','longitude']
data_all_temp = data_all.loc[data.index]
cat_cols = data_all_temp[cols]
data = pd.concat([data, cat_cols], axis=1)
```

```python
data
```

```python
cols = ['latitude','longitude']  + [col for col in data if col != ['latitude','longitude']]
data = data[cols]
```

```python
data.to_csv('listings_cleaned_with_counts.csv')
```
